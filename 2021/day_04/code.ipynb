# Puzzle 1

```python
file = open('i')
numbers = next(file).split(',')
boards = [[[*map(int,next(file).split())] for r in range(5)] for b in file]
boards += [[*map(list,zip(*board))] for board in boards]

for number in numbers:
    for board in boards:
        for row in board:
            if (number:=int(number)) in row:
                row.remove(number)
                if not row:
                    print(sum(map(sum,board))*number)
```
